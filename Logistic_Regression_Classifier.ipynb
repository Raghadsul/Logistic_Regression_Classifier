# Logistic Regression Classifier

```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

### Load Dataset

```python
df = pd.read_csv('datasets/iris.csv')
df.head()
```

### Plot Dataset

```python
ax = sns.scatterplot(x="sepal.length", y="sepal.width", hue='variety', size="variety"
                     , sizes=(40, 100), hue_norm=(0, 7), legend="full", data=df)
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
plt.legend()
plt.show()
```

```python
ax = sns.scatterplot(x="petal.length", y="petal.width", hue='variety', size="variety"
                     , sizes=(40, 100), hue_norm=(0, 7), legend="full", data=df)
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.legend()
plt.show()
```

## Training and Testing

### Split Dataset using Cross-validation for (sepal.length, sepal.width) featuers

```python
X = df.iloc[:, 0:2]
y = df.iloc[:, 4]
model = LogisticRegression()
scores = cross_val_score(model, X, y, cv=10,scoring='accuracy')
```

```python
print(scores)
print(scores.mean())
```

### Split Dataset using Cross-validation for all featuers

```python
X = df.iloc[:, 0:4]
y = df.iloc[:, 4]
model = LogisticRegression()
scores = cross_val_score(model, X, y, cv=10,scoring='accuracy')
```

```python
print(scores)
print(scores.mean())
```

### Cross-validation for all features with square of sepal.length and sepal.width

```python
X['v1squared']= [n**2 for n in df['sepal.length']]
X['v2squared']= [n**2 for n in df['sepal.width']]
X['v3pow2']= [n**2 for n in df['petal.length']]
X['v4pow2']= [n**2 for n in df['petal.width']]
model = LogisticRegression()
scores = cross_val_score(model, X,y, cv=10,scoring='accuracy')
```

```python
print(scores)
print(scores.mean())
```
